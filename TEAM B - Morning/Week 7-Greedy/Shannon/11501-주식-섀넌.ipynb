# 문제 읽기
- 1. 주식을 산다
- 2. 원하는 만큼 가지고 있는 주식을 판다
- 3. 아무것도 안한다
- 최대 이익이 얼마나 되는지 계산하기
- 입력:
    - 1: 테스트 케이스 수 t
        - 1 : 날의 수 n
        - 2 : 날 별 주가

# 문제 전략
- 최소르 보자마자 왤케 dfs로 풀고 싶지?
- 올라갈때 샀다가, 내려가려고 할때 팔기

```python
t = int(input())
for i in range(t):
    n = int(input())
    arr= list(map(int, input().split()))

    max_value = arr[n-1]
    target = [0]*n
    for i in range(n-1,-1,-1):
        max_value = max(max_value, arr[i])
        target[i] = max_value

    result = 0
    for i in range(n):
        result += max(0,target[i]-arr[i])
    print(result)
```
