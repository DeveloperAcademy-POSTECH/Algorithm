# 문제 해석
- 양방향 순환 큐 : 원소 뽑아내기 
- 첫번째 원소 뽑아내기 -> pop(0)
- 왼쪽으로 한칸 이동 -> append(pop(0)
- 오른쪽으로 한칸 이동 -> pop(), for문으로 한칸씩 뒤로가게 하고,0번째에 더하기
- 큐에 처음에 포함되었던 수 N 주어짐, 지밍이가 뽑아내려고 하는 원소의 위치가 주어짐 : 가장 처음 큐에서의 위치
- 그 원소를 주어진 순서대로 뽑아드는데 2,3번 원소의 최솟값

- 입력 : 큐의 크기 N 뽑아내려고 하는 수의 개수 M- N보다 작거나 같은 자연수, 둘째줄에는 지민이가 뽑아내려고 하는 수의 위치(인덱스)

# 전략
-전체 개수를 기준으로 뒤로 갈지(오른쪽으로 갈지) (왼쪽으로 갈지 결정해야 함)
- 한 턴을 돌때마다 인덱스 번호가 업데이트 되어야 함 

```python
n,m = map(int, input().split(' '))
indexList = list(map(int, input().split(' ')))

half = int(n/2)
second =0
third =0 

for i in range(m) :
    if indexList[i] != 1 or indexList[i] != n:
        if indexList[i] <= half :
            while indexList[i] != 1:
                second += 1
                indexList[i] -= 1
                for j in range(i+1,m):
                    indexList[j] -= 1
                    if indexList[j]>n :
                        indexList[j] -= n
                    elif indexList[j] <1 :
                        indexList[j] += n

        elif indexList[i] > half :
            while indexList[i] != n :
                third += 1
                indexList[i] += 1
                for j in range(i+1, m):
                    indexList[j] += 1  
                    if indexList[j]>n :
                        indexList[j] -= n
                    elif indexList[j] <1 :
                        indexList[j] += n      

    else :
        break

print(second + third)



```

```python
# Pop 말고 deque를 쓰는 이유는 pop(0)은 시간 복잡도가 엄청나게 늘어난다
n,m = map(int, input().split())
indexList = list(map(int, input().split()))
numList = list([i for i in range(1,n+1)])

half = int(n/2)
second =0
third =0 

for i in indexList :
    index = numList.index(indexList[i])
    if index <= half :
        for i in range(index) :
            x = numList.pop(0)
            numList.append(x)
    else :
        for i in range(index) :
            x = numList.pop()
            numList.insert(0,x)
    




    else :
        break

print(second + third)
```

```python
# import sys
# input = sys.stdin.readline 
from collections import deque
 #원소의 개수, 뽑아내는 횟수

n,m = map(int, input().split())
de_que = deque([i for i in range(1,n+1)])#1부터 n가지의원소를 삽입
targets = list(map(int,input().split()))#뽑아낼 원소 목록
cnt = 0 # 회전 연산 수행 횟수

x=0

for target in targets : # 뽑아낼 원소를 하나씩 확인하며
    #내가 생각할 때는 다른 요소는 맞췄는데, 
    index = de_que.index(target) # 덱에서 해당 원소의 위치를 찾기 : 해당 요소의 인덱스가 변할 수 있어서 인덱스와 타겟을 따로 두었던 
    if index <= len(de_que)//2 : # 왼쪽에서 돌리는게 더 빠른 경우
        for i in range(index) : # 해당 위치까지 왼쪽으로 계속 빼기 : 이거 자체가 회전 연산 
            x = de_que.popleft()
            de_que.append(x)
            cnt += 1
    else : # 오른쪽에서 돌리는게 더 빠른 경우
        for i in range(len(de_que) - index) : #회전연산 반복 수행
            x = de_que.pop()
            de_que.appendleft(x)
            cnt += 1
    de_que.popleft()

print(cnt)
```

```python
#이게 좀더 직관적
from collections import deque
import sys
input = sys.stdin.readline

n, m = map(int, input().split())    # 큐의 크기 n과 뽑아내려고 하는 수의 개수 m을 입력값으로 받기
position = list(map(int, input().split()))  # 뽑아내려는 수의 위치를 입력값으로 받기
dq = deque([i for i in range(1, n+1)])  # deque([1, 2, 3,...,n])

count = 0   # 2, 3번 수행하면 카운트 올리기
for i in position:  # 뽑아내려는 수의 위치 하나씩 반복문 돌리기
    while True:     # 뽑을 때까지 계속 돌리기
        if dq[0] == i:  # dq의 첫인덱스가 뽑아내려는 수의 위치와 같다면 1번 수행
            dq.popleft()
            break
        else:
            if dq.index(i) < len(dq)/2:  # 뽑아내려는 수의 위치 인덱스가 dq의 길이를 반으로 나눈것보다 작을때는 왼쪽으로 움직여야 최소
                while dq[0] != i:   # dq의 첫번째 인덱스가 i와 같아질때까지 반복
                    dq.append(dq.popleft())  
                    count += 1
            else:   # 뽑아내려는 수의 위치 인덱스가 dq의 길이를 반으로 나눈것보다 클때는 오른쪽으로 움직여야 최소
                while dq[0] != i:
                    dq.appendleft(dq.pop())  
                    count += 1
print(count)
```

```python
# 내가 놓친 포인트 : 꼭 앞에서만 뽑아낼 수 있음 
from collections import deque
import sys
input = sys.stdin.readline

n, m = map(int,input().strip().split())
position = list(map(int, input().split()))
dq = deque([i for i in range(1,n+1)])

count = 0
for i in position:
    while True:
        if dq[0] ==i :
            dq.popleft()
            break
        else :
            if dq.index(i) <len(dq)/2 :
                while dq[0]!= i :
                    dq.append(dq.popleft())
                    count += 1
            else:
                while dq[0]!= i :
                    dq.appendleft(dq.pop())
                    count += 1
print(count)
```
