##문제 전략
- 사실 어떻게 문제를 풀어야 하는지 잊어버렸지만,,, 음
- 일단 무게가 적은 순으로 sort를 함 : 힙을 사용하면 좋을 듯?
- 그리고 무게가 될때까지 합을 더하는 식으로, 이게 7까지 들어갈 수 있으니까 7까지 리스트를 만들어 놓고, 물건의 개수만큼 턴을 돌리면 될듯?
- 만약에 4가 들어왔을때, 4가 더해져서 가능한 것에 모두 무게를 넣기


```python
n,k = map(int, input().split())
dp= [[0]*(k+1) for i in range(n+1)]

for i in range(1,n+1):
    weight, value = map(int, input().split())
    for j in range(1,k+1):
        if j < weight:
            dp[i][j] = dp[i-1][j]
        else:
            dp[i][j] = max(dp[i-1][j], dp[i-1][j-weight]+value)

print(dp[n][k])
```
