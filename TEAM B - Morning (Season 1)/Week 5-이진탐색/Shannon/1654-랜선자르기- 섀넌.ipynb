# 문제 읽기
- N개의 랜선을 만들어야 함
- 이미 K개의 랜선을 가지고 있음, K개의 랜선은 길이가 제각각
- N개의 같은 길이의 랜선을 만들고 싶음 -> K개의 랜선을 잘라서 만들어야 함
- N개보다 많이 만드는 것도 N개에 포함됨, 만들 수 있는 최대 랜선 길이는?
- 입력
    - 이미 가지고 있는 랜선의 개수 K(만 이하), 필요한 랜선의 개수 N(백만 이하)
    - K의 길이
# 문제 전략
- 완전 딱봐도 이진탐색 문제
- 간격을 어떻게 해야할까,
- 중앙값 설정 : 0부터 가장 작은 길이의 중앙값?

```python
import sys
# input = sys.stdin.readline

k,n = map(int, input().split())
K = []
for i in range(k):
    K.append(int(input()))

end = max(K)
start = 1
result = 0

while start <= end :
    counts = 0
    mid = (start + end)//2
    for i in K:
        counts += i//mid
    if counts >= n :
        start = mid +1
        result = mid
    else:
        end = mid -1

print(result)



```

```python
k,n = map(int, input().split())
K = []
for i in range(k):
    K.append(int(input()))
start, end = 1, max(K)

while start <= end: #적절한 랜선의 길이를 찾는 알고리즘
    mid = (start + end) // 2 #중간 위치
    lines = 0 #랜선 수
    for i in K:
        lines += i // mid #분할 된 랜선 수

    if lines >= n: #랜선의 개수가 분기점
        start = mid + 1
    else:
        end = mid - 1
print(end)
```
