#문제 읽기
- 숫자카드는 정수가 하나 적혀져 있는 카드
- N개 가지고 있음, 정수 M개 가지고 있을대
- 몇개를 가지고 잇는지


```python
n = int(input())
cards = list(map(int, input().split()))
m = int(input())
numbers = list(map(int,input().split()))
counts = {}
for i in range(m):
    counts[numbers[i]] = 0

for i in range(n):
    if cards[i] in counts:
        counts[cards[i]] += 1

for i in range(m):
    print(counts[numbers[i]], end =' ')
```
