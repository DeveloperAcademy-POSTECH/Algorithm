- 문제 해석
    - 첫째줄에는 기준 수의 개수 N
    - 둘째줄에는 기준 수들 : N개의 정수
    - 셋째줄에는 있는지 확인할 수의 수 : M
    - 넷째줄에는 잇는지 확인할 수 : M개의 정수
- 문제 풀이
    : 어떤것이 있고, 없고는 set으로 확인하면 됨

```python
import sys

N = int(sys.stdin.readline())
normSet = set(map(int, sys.stdin.readline().strip(' ')))
M = int(sys.stdin.readline())
checkNum = list(map(int, sys.stdin.readline().strip(' ')))

for num in checkNum :
    if num in normSet :
        print('1')
    else :
        print('0')
```
