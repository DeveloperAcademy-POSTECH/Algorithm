# 문제 읽기
- 입력
    - 1 : N(나무 수), M(집으로 가져가려는 나무 길이)
    - 2 : 나무의 높이, 나무 높이 합은 항상 M보다 큼
- M미터가 필요,
- 목재 절단기 높이 H를 지정해야함, 그 다음 한줄에 연속해 있는 나무를 모두 절단해 버림
- 높이가 H보다 큰 나무는 H위 부분이 잘릴 것이고, 낮은 나무는 안잘림
- 그리고 남은 높이를 자르고 가져감

# 문제 전략
- 완전 이진탐색 문제

```python
n, m = map(int, input().split())
trees = list(map(int, input().split()))
start = 1
end = max(trees)
result = 0
while start <= end :
    mid = (start+end)//2
    counts = 0
    for i in trees :
        if i > mid :
            counts += i-mid
    if counts >= m :
        start = mid +1
        result = mid
    else :
        end = mid -1

print(result)
```
