## 문제 읽기
- N개의 컴퓨터
- A가 B를 신뢰할때, B해킹 시 A해킹 가능
- 한번에 가장 많은 컴퓨터를 해킹할 수 있는 컴퓨터 번호를 출력하는 프로그램
- 입력
    - N: 컴퓨터 수, M : 연결관계 수
    - M
        - A,B ㅎ형식 : A,B 신뢰

# 문제 전략
- 한방향 BFS, DFS

```python
from collections import deque
n, m = map(int, input().split())
adj = [[] for i in range(n+1)]
for i in range(m):
    x,y = map(int,input().split())
    adj[y].append(x)
visited = [False]*(n+1)
count = [0]*(n+1)

def bfs(v):
    q = deque([v])
    visited = [False]*(n+1)
    visited[v]= True
    count = 1
    while q :
        v = q.popleft()
        for e in adj[v]:
            if not visited[e]:
                q.append(e)
                visited[e] = True
                count += 1
    return count

result = []
max_value = -1
for i in range(1,n+1):
    c = bfs(i)
    if c>max_value:
        result = [i]
        max_value = c
    elif c == max_value:
        result.append(i)
        max_value = c

print(*result)

```
