# 문제 읽기
- 네트워크에 연결된 컴퓨터가 웜 바이러스에 걸림
- 1번 컴퓨터가 웜바이러스에 걸렸을때 웜 바이러스에 걸리는 컴퓨터의 수를 출력
- 입력
    - 1 : 컴퓨터 수(100이하)
    - 네트워크에 연결된 컴퓨터 쌍 수 
- 출력 : 컴퓨터 웜 바이러스에 걸리는 컴퓨터의 수
## 전략
- 예전에 풀었던 사람 네트워크 문제랑 비슷한것 같기도 하고 : find&union
- 네트워크라고 하니까 bfs로도 풀 수 있을듯?, 그리고 상호간의 방향성이 없으니까 둘다 담아야 할듯

```python
from collections import deque
n = int(input())
m = int(input())
array = [[] for i in range(n+1)]
for i in range(m):
    x, y = map(int, input().split())
    array[x].append(y)
    array[y].append(x)

visited = []
count = 1

def bfs():
    now_visit = deque([1])
    if now_visit :
        node = now_visit.popleft()
    for node not in visited :
        visited.append(node)
        count += 1
        now_visit.append(array[node])
    return count 


```

```python
n = int(input())
m = int(input())
adj = [[]for i in range(n+1)]
visited = [False]*(n+1)
count = 0

for i in range(m):
    x,y = map(int, input().split())
    adj[x].append(y)
    adj[y].append(x)

def dfs(now_pos):
    global count
    count += 1
    visited[now_pos]=True
    for next_pos in adj[now_pos]:
        if not visited[next_pos]:
            dfs(next_pos)

dfs(1)
print(count-1)
```

# 문제 읽기
- 바이러스가 퍼진다. 
- 연결되어 있으면 다 퍼진다. 
- 1번 컴퓨터가 바이러스에 걸릴때, 1번 컴퓨터를 통해 바이러스에 걸리게 되는 컴퓨터의 수를 출력하셈
- 입력
    - 1: 컴퓨터 수
    - 2 : 연결되어있는 네트워크쌍 수 
# 문제 전략
- 이건 최소가 아니니 더 편한 dfs로 풀어보자 
- 양방향에 대한 내용이 안나와 있으니 일단 양방향으로 데이터 저장하자
- 1번이랑 연결되어있는 걸 카운팅 해야 함

```python
n = int(input())
m = int(input())
arr = [[] for i in range(n+1)]
for i in range(m):
    a, b = map(int, input().split())
    arr[a].append(b)
    arr[b].append(a)

visited = [False]*(n+1)
count = 0

def dfs(v):
    global count
    visited[v]= True 
    count += 1
    for i in arr[v]:
        if not visited[i] :
            dfs(i)
dfs(1)
print(count-1)
```
