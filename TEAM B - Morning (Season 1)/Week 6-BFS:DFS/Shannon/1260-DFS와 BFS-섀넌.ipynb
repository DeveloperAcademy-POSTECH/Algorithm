- 더 이상 방문할 수 있는 점이 없는 경우 종료
- 간선은 양방향
- 입력
    - 1 : N 정점의 개수 : 1~1000, M 간선의 개수: 1~1만, 탐색을 시작할 번호 V
    - M
        - 두 정점의 번호
- 출력
    - 첫째줄에 DFS를 수행한 결과, 다음줄에는 BFS를 수행한 결과를 출력,
# 문제 전략
    - 입력을 그래프로 만들어야 하는데, 딕셔너리로 만들라고 했으까, 딕셔너리로 만든 다음에 extend할때  한번씩 sort를 해주자
