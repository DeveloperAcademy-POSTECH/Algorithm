#전략
- 첫번째 줄에는 K -> 그 이후에 K개의 줄 -> 정수가 0일 경우에는 가장 최근에 쓴 수를 지우기, 아닐 경우 수를 씀


```python
K = int(input())
numList = []
for _ in range(K) :
    num = int(input())
    if num == 0 :
        numList.pop()
    else :
        numList.append(num)
print(sum(numList))
```
