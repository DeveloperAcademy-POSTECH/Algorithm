```python
#인덱스에 따라 오큰수를 담아야 함 -> 인덱스도 함께 기억해야 함
# 내림차순으로 리스트가 주어질때는 계속 스택에 담겨야 하고, 역순이 주어질때 오큰수가 담아짐 
N= int(input())
numList = list(map(int, input().split(' ')))
stack =[]
NGE =[-1] # 오큰수 배열

for i in range(N) :
    x = numList[i]
    if len(stack) <1 or stack[-1][0] >= x :# 스택에 한개도 없거나, 맨 끝에 있는 수가 x보다 크다면 
        stack.append((x,i))
    else : 
        while len(stack) >0 : 
            previous, index = stack.pop() # 계속 내림차순으로 왔다면, x보다 크기 전까지 계속 오큰수는 x인 부분을 찾아주는 것이 필요 
            if previous >= x : # x보다 큰 previous를 만난다면 다시 삽입
                stack.append((previous,index))
                break
            else:
                NGE[index] = x # 오큰수 기록 

for x in NGE :
    print(x,end = '')
        


```

```python
# import sys
# input = sys.stdin.readline

n = int(input())
numList = list(map(int, input().split(' ')))
stack =[]
NGE = [-1] * n # 오큰수 배열 - n개모두 -1일 수 있으니까 -1로 초기화  

for i in range(n):
    x = numList[i] #하나씩 수 확인 
    if len(stack) == 0 #or stack[-1][0] >= x :
        stack.append((x,i)) # (수,인덱스) 형태로 삽입 
    else : 
        while len(stack) >0 :
            previous, index = stack.pop()
            if previous >=x : #크거나 같은 이전 원소를 만났다면 다시 삽입
                stack.append((previous,index))
                break
            else :
                NGE[index] = x #오큰수 기록
        stack.append((x,i)) 

for x in NGE :
    print(x,end=' ')

```
