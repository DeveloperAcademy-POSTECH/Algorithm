# 문제 읽기
- 수열 A가 주어졌을때 가장 긴 증가하는 순열
# 문제 전략
- 배낭문제 같은 동적 계획볍으로 풀어볼까?

```python
n = int(input())
numbers = list(map(int, input().split()))
dp = [1]*n
for i in range(1,n):
    for j in range(i):
        if numbers[j]<numbers[i] :
            dp[i] = max(dp[i],dp[j]+1)

print(max(dp))
```
