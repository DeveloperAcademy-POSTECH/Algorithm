# 문제 읽기
- X가 3으로 나누어 떨어지면, 3으로 나눈다.
- X가 2로 나누어 떨어지면, 2로 나눈다.
- 1을 뺀다.
- 연산을 사용하는 최소 횟수

```python
n = int(input())
dp = [0]*(n+1)
for i in range(2,n+1):
    d[i] = d[i-1] +1
    if i %3 == 0 :
        d[i] = min(d[i],d[i//3]+1)
    if i%2 == 0:
        d[i] = min(d[i],d[i//2]+1)
print(d[n])
```
