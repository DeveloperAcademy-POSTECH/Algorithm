# 문제 읽기
- 크레인이 n대 있고, 1분에 박스를 하나씩 배에 실을 수 있음
- 각 크레인의 무게 제한
- 입력
    - 1: n
    - 2 : n개의 크레인의 무게 제함
    - 3 : 옮겨야 할 짐
    - 4: 각 짐의 무게

# 문제 전략
- 핵심은 최대한 동일한 양의 무게를 옮겨야 한다는 것
- n이 50으로 작으니까 그냥 전체 리스트를 만들고,
- 박스의 무게가 크레인보다 작은지 확인 : 가동 가능한지에 대한 False 리스트 만들기
- True인 애들 중에서 작은 값을 더해주기
- 만약에 모두보다 큰 짐이 있다면 -1을 프린트하고 종료


```python
n = int(input())
crain = list(map(int, input().split()))
m = int(input())
goods = list(map(int, input().split()))
all_count = [0]*51
crain.sort()
cant = True

for i in range(m):
    all_can = [False]*51
    for j in range(n):
        if goods[i] <= crain[j]:
            all_can[j] = True
        else :
            cant = False
            break
    all_can_index = []
    for k in range(1,51):
        if all_can[i] == True:
            all_can_index.append(i)
    min_index = 1e6
    for l in range(len(all_can_index)):


    index_b = min(all_count[i] for i range(51) if all_can[i] == True)
    index_a = all_count.index(index_b)
    all_count[index_a] += 1

if cant :
    print(max(all_count))
else :
    print(-1)
```

```python
#시간초과
import sys

n = int(input())
cranes = list(map(int, input().split()))
m = int(input())
boxs = list(map(int, input().split()))

cranes.sort(reverse = True)
boxs.sort(reverse = True)
if cranes[0]<boxs[0] :
    print(-1)
    sys.exit()
else :
    time = 0
    while boxs:
        if not boxs:
            break
        for crane in cranes:
            for box in boxs:
                if crane >= box:
                    boxs.remove(box)
                    break

        time += 1

    print(time)







```

```python
import sys

n = int(input())
cranes = list(map(int, input().split()))
m = int(input())
boxes = list(map(int, input().split()))

cranes.sort(reverse = True)
boxes.sort(reverse = True)
positions = [0]*n
checked = [False]*m
result = 0
count = 0
if cranes[0]<boxes[0] :
    print(-1)
    sys.exit()

while True :
    if count == len(boxes):
        break
    for i in range(n):
        while positions[i]<len(boxes):
            if not checked[positions[i]] and cranes[i] >= boxes[positions[i]]:
                checked[positions[i]] = True
                positions[i] += 1
                count += 1
                break
            positions[i] += 1
    result += 1
print(result)
```
