# 문제 읽기
- 알파벳 순으로 정렬하려고 함
- 사전순이 가장 위
- 책 하나를 빼서 가장 위에 놓기
# 문제 전략
- 뒤에서부터 훑어서 만약에 n-1이 n보다 뒤에 있다면 앞으로 이동

```python
n = int(input())
books = []
for i in range(n):
    books.append(int(input()))

books.reverse()
count = 0

for i in range(n,1,-1) :
    x = books.index(i)
    y = books.index(i-1)
    if x > y :
        del books[y]
        books.append(i-1)
        count += 1

print(count)


```
